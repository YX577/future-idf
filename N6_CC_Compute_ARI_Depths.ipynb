```r
library(extRemes)
library(glue)
```

```r
### MODELS LOCA ###

modelloca_hist <- c("IPSL-CM5A-MR", "ACCESS1-3", "MRI-CGCM3", "HadGEM2-CC", "CNRM-CM5", "CanESM2", 
"ACCESS1-0", "HadGEM2-AO", "CSIRO-Mk3-6-0", "IPSL-CM5A-LR", "FGOALS-g2", "GISS-E2-H",
"inmcm4", "CESM1-CAM5", "HadGEM2-ES", "GFDL-ESM2G", "CMCC-CMS", "MIROC-ESM-CHEM", "CESM1-BGC", 
"GFDL-ESM2M", "CMCC-CM", "GISS-E2-R", "bcc-csm1-1-m", "CCSM4", "EC-EARTH", 
"bcc-csm1-1", "MPI-ESM-MR", "MIROC5", "NorESM1-M", "MPI-ESM-LR", "GFDL-CM3", "MIROC-ESM")
```

```r
dataset <- c('loca')
scenario <- c('hist', 'rcp45-1', 'rcp45-2', 'rcp85-1', 'rcp85-2')
bound <- c(" ")
```

```r
for (d in dataset) {
    
        for (s in scenario) {
            
            for (bnn in bound){
        
       modelex = modelloca_hist
        #message(glue('output/missing/returnlevels/{d}/{s}/{d}.{s}.{m}.csv'))
        for (m in modelex) {
        infile <- glue('gcm_data/params/{s}/{d}.{s}.{m}.csv')
        params <- read.csv(file=infile, header=T, sep=',')
        message(glue('Dataset: {d}, Scenario: {s}, Model: {m}'))
            
        if (bnn == 'U_'){
            df = subset(params, select = -c(L_loc, loc, L_scale, scale, L_shape,shape))
        } else if (bnn == 'L_'){
            df = subset(params, select = -c(U_loc, loc, U_scale, scale, U_shape,shape) )
        } else if (bnn == ' '){
            df = subset(params, select = -c(U_loc, L_loc, U_scale, L_scale, U_shape, L_shape))
        }
        params <- t(df)
        model.name <- params["models", ]
        #message(" ", model.name)

        f = (length(model.name))
        rl.cals <- data.frame(matrix(ncol = 7, nrow =length(model.name) ))
        colnames(rl.cals) <- c("models", "2-yr", "5-yr",
                                         "10-yr", "25-yr", "50-yr",
                                         "100-yr")
        rl.cals[1] <- as.vector(model.name)

        for (model in 1:f){

            pmodel <- params[, (model)]
            pmodel <- pmodel[!is.na(pmodel)]
            tryCatch({
            rl.cals.single <- as.vector(rlevd(period=c(2,5,10,25,50,100), 
              loc = as.double(pmodel[3]), scale=as.double(pmodel[4]), shape=as.double(pmodel[5])))
            rl.cals[model, c(2:7)] <- rl.cals.single
            }, error = function(e){})
        }
            
        mainDir = glue('gcm_data/returnlevels/')
        subDir2 = glue('{s}')
        dir.create(file.path(mainDir, subDir2), showWarnings = FALSE)
        write.csv(rl.cals, glue('gcm_data/returnlevels/{s}/{d}.{s}.{m}.csv'))
    }
}
}    
}
```
